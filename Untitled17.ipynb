```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# one hot encoding
pd.get_dummies(df,columns=['Geography', 'Gender'], drop_first=True)
```

```python
x = df.drop(['Exited'], axis=1)
y = df['Exited']
x = pd.get_dummies(x,columns=['Geography', 'Gender'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state =42)
```

```python
#Standard Scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x.shape
```

```python
#Building - ANN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.activations import relu, sigmoid
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
```

```python
#Input Layer
model.add(Dense(units=11, activation='relu'))
model.add(Dropout(0.2))
```

```python
# 1st Hidden Layer
model.add(Dense(units =7, activation='relu'))
model.add(Dropout(0.2))
```

```python
# 2ND LAYER
model.add(Dense(units =6, activation='relu'))
model.add(Dropout(0.2))
```

```python
# OUPUT LAYER
model.add(Dense(units =1, activation='sigmoid'))
model.add(Dropout(0.2))
```

```python
early=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=0,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
# # from keras.model loaded sequntial
# # created model as Sequential
# # from keras.layers
# # 1st hidden layer
# # 2nd hidden layer
# # output layer
# # # we complete model by setting some optimier and loss function and accuracy metrices
# we train for training  data -> x_train, x_test
# _________________________________
# early stopping, dropout, validation_split
# prediction done
# # evaluation done
#dropout layer used to control neurons or hidden layer they make neurons dead by giving value
#earlystopping stops training the model onece if the loss is not changing - add at the training time

# plot compare training loss
# plot compare validation loss
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=10, epochs=100,validation_split=0.3, callbacks=[early])
```

```python
y_pred_prob = model.predict(x_test)
```

```python
y_pred = (y_pred_prob>0.5).astype(int)
```

```python
#Evaluation
from sklearn.metrics import accuracy_score, classification_report
```

```python
y_pred_train = (model.predict(x_train)).astype(int)
```

```python
y_pred_train
```

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
classification_report(y_test, y_pred)
```

## Summary of Key Findings and Feature Engineering Opportunities

### Subtask:
Summarize the key findings and identify potential feature engineering opportunities from this analysis.

### Key Findings:

1.  **Exited (Churn)**:
    *   Approximately 20% of customers churned, indicating an imbalanced dataset, which should be considered during model training and evaluation.
2.  **CreditScore**:
    *   Customers with lower credit scores seem to have a slightly higher propensity to churn, though the overlap is significant.
3.  **Age**:
    *   A higher proportion of customers in the middle age range (e.g., 40-60) tend to churn compared to younger or older customers. This suggests age might be a significant factor.
4.  **Tenure**:
    *   The distribution of tenure for churned and non-churned customers is quite similar, suggesting tenure alone might not be a strong predictor.
5.  **Balance**:
    *   A large number of customers have zero balance. Among customers with a balance, those who churned often have higher balances. This could indicate customers with more assets are also more willing to move their money to other banks, or perhaps dissatisfaction among those with significant funds.
6.  **EstimatedSalary**:
    *   The estimated salary distribution is fairly uniform for both churned and non-churned customers, implying it might not be a strong differentiating factor.
7.  **Geography**:
    *   Germany has a disproportionately higher churn rate compared to France and Spain, even though France has the largest customer base. This suggests regional differences are important.
8.  **Gender**:
    *   Female customers appear to have a slightly higher churn rate than male customers.
9.  **NumOfProducts**:
    *   Customers with 3 or 4 products show a very high churn rate, while those with 1 or 2 products have lower churn. This could indicate dissatisfaction or complexity with multiple products.
10. **HasCrCard**:
    *   Having a credit card does not seem to significantly impact churn, with similar proportions among churned and non-churned customers.
11. **IsActiveMember**:
    *   Inactive members have a considerably higher churn rate than active members, which is an expected but important finding.

### Potential Feature Engineering Opportunities:

1.  **Age Bins/Categories**: Given the churn pattern observed in the 'Age' distribution, creating age categories (e.g., young, middle-aged, senior) could capture non-linear relationships.
2.  **Balance Status**: Create a binary feature `HasZeroBalance` (or similar) to differentiate customers with zero balance from those with a positive balance, as 'Balance' has a large number of zero values and might behave differently.
3.  **Credit Score Categories**: Binning `CreditScore` into categories (e.g., excellent, good, fair, poor) might better capture its relationship with churn.
4.  **Interaction Features**: Explore interaction terms, such as `Age` * `NumOfProducts`, or `Balance` * `IsActiveMember`, to see if combined effects are more predictive.
5.  **Ratio Features**: Create ratios like `Balance/EstimatedSalary` to understand financial behavior relative to income.
6.  **Product Complexity Score**: A score based on `NumOfProducts` that penalizes higher product counts could be beneficial, given the high churn for customers with many products.
7.  **Geography-specific interactions**: Since Germany has higher churn, interaction features with `Geography_Germany` could be useful to model specific regional effects.
